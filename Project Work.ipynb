7. What is the 'Wkly_Study_Hours' of the overall best student/students?


```python
# To get the overall best student, we need create a new column called 'Average_score'
df['Average_score'] = (df['Math_Score'] + df['Reading_Score'] + df['Writing_Score'])/ 3 # Avearge of the three subjects
```

```python
df['Average_score'] = round(df['Average_score'], 1) # Rounding df['Average_score'] to 1 decimal place
```

```python
best_average = df['Average_score'] == 100 # number of students that scored 100 marks on average
best_average.sum()
```
